### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_predict
from pmdarima.arima import auto_arima
from sklearn.metrics import mean_squared_error

from library.functions import jlcap_optimize_arima
```

### Notebook Settings

```python
pd.set_option('display.max_rows', 500)
```

### Load the data

```python
y = pd.read_csv('data/data_pre-processed.csv')
y.set_index('date_forward', inplace=True)

y_diff = pd.read_csv('data/data_stationary.csv')
y_diff.set_index('date_forward', inplace=True)

y_avg = pd.read_csv('data/data_avg_pre-processed.csv')
y_avg.set_index('date_forward', inplace=True)
```

```python
y.head()
```

```python
y.index = pd.DatetimeIndex(y.index, freq="D")
y_diff.index = pd.DatetimeIndex(y_diff.index, freq="D")
y_avg.index = pd.DatetimeIndex(y_avg.index, freq="D")
y.index
```

```python
y_diff.head()
```

```python
y_avg.head()
```

### Split Data into Training and Testing Sets

```python
len(y)
```

```python
split = int(np.floor(len(y)*.8))
split
```

```python
y_train = y[:split]
y_test = y[split:]
```

```python
y_train.shape, y_test.shape
```

**y_avg**

```python
y_avg_train = y_avg[:split]
y_avg_test = y_avg[split:]
y_avg_train.shape, y_avg_test.shape
```

## Model 1: ARIMA

### Finding ARIMA Parameters

```python
ps = range(0,8)
qs = range(0,8)

params = jlcap_optimize_arima(y_train, ps, qs)
params
```

```python
ps = range(0,8)
qs = range(0,8)

params_avg = jlcap_optimize_arima(y_avg_train, ps, qs, d=0)
params_avg
```

### Fitting a Model

```python
mod = ARIMA(y_train, order=(4,1,7))
res = mod.fit()

print(res.summary())
```

```python
res.plot_diagnostics()
plt.show()
```

```python
mod_avg = ARIMA(y_avg_train, order=(3,0,3))
res_avg = mod_avg.fit()

print(res_avg.summary())
```

```python
res_avg.plot_diagnostics()
plt.show()
```

**Predictions**

```python
fig, ax = plt.subplots(figsize=(16, 8))

fig = plot_predict(res, start="2021-03-12", end="2021-05-24", ax=ax)
ax.plot(y, label='Actual')
legend = ax.legend(loc="upper left")
```

```python
predictions = res.forecast(steps=len(y_test))
predictions.rename('bg_fasting', inplace=True)
```

```python
y_pred = pd.DataFrame(predictions)
y_pred.tail()
```

```python
res.resid.plot()
```

```python
rmse = mean_squared_error(y_test, y_pred, squared=False)
rmse
```

```python
res_conf_int = res.conf_int()
res_conf_int
```

```python
# Plot the points and the forecasts
training_days = pd.Series(y_train.index)
testing_days = pd.Series(y_test.index)

plt.figure(figsize=(14,7))

plt.plot(training_days, y_train, alpha=0.75, label='Training Data')
plt.plot(testing_days, y_pred, alpha=0.75, label='Predictions')  # Forecasts
plt.plot(testing_days, y_test,
            alpha=0.4, linestyle='--', label='Actual/Test Data')  # Test data
# plt.fill_between(testing_days, res_conf_int[:, 0], res_conf_int[:, 1], alpha=0.1, color='b')

plt.title("GDB forecasts using ARIMA")
plt.xlabel("Date")
plt.ylabel("Blood Glucose Reading (mg/dL)")
plt.legend(loc='upper left')

plt.savefig('vis/arima_pred_vs_test.png', bbox_inches="tight")
plt.show()
```

**y_avg_train**

```python
fig, ax = plt.subplots(figsize=(16, 8))

fig = plot_predict(res_avg, start="2021-03-12", end="2021-05-24", ax=ax)
ax.plot(y_avg, label='Actual')
legend = ax.legend(loc="upper left")
```

```python
predictions_avg = res_avg.forecast(steps=len(y_avg_test))
predictions_avg.rename('bg_avg', inplace=True)
```

```python
y_pred_avg = pd.DataFrame(predictions_avg)
y_pred_avg.tail()
```

```python
res_avg.resid.plot()
```

```python
rmse_avg = mean_squared_error(y_avg_test, y_pred_avg, squared=False)
rmse_avg
```

```python
res_conf_int_avg = res_avg.conf_int()
res_conf_int_avg
```

```python
plt.figure(figsize=(14,7))

plt.plot(training_days, y_avg_train, alpha=0.75, label='Training Data')
plt.plot(testing_days, y_pred_avg, alpha=0.75, label='Predictions')  # Forecasts
plt.plot(testing_days, y_avg_test,
            alpha=0.4, linestyle='--', label='Actual/Test Data')  # Test data
# plt.fill_between(testing_days, res_conf_int[:, 0], res_conf_int[:, 1], alpha=0.1, color='b')

plt.title("GDB forecasts using ARIMA")
plt.xlabel("Date")
plt.ylabel("Blood Glucose Reading (mg/dL)")
plt.legend(loc='upper left')

plt.savefig('vis/arima_avg_pred_vs_test.png', bbox_inches="tight")
plt.show()
```

## Model 2: auto_arima

```python
# Fit a simple auto_arima model
mod_auto = auto_arima(y_train, start_p=0, start_q=2,
                     max_p=10, max_q=10, d=1,
                     stepwise=True, suppress_warnings=True,
                     error_action='ignore')

mod_auto.order
```

```python
mod_auto.summary()
```

```python
# Create predictions for the future, evaluate on test
preds, conf_int = mod_auto.predict(n_periods=y_test.shape[0], return_conf_int=True)

# Print the error:
print("Test RMSE: %.3f" % np.sqrt(mean_squared_error(y_test, preds)))
```

```python
# Plot the points and the forecasts
training_days = pd.Series(y_train.index)
testing_days = pd.Series(y_test.index)

plt.figure(figsize=(14,7))

plt.plot(training_days, y_train, alpha=0.75, label='Training Data')
plt.plot(testing_days, preds, alpha=0.75, label='Predictions')  # Forecasts
plt.plot(testing_days, y_test,
            alpha=0.4, linestyle='--', label='Actual/Test Data')  # Test data
plt.fill_between(testing_days,
                 conf_int[:, 0], conf_int[:, 1],
                 alpha=0.1, color='b')

plt.title("GDB forecasts using auto_arima")
plt.xlabel("Date")
plt.ylabel("Blood Glucose Reading (mg/dL)")
plt.legend(loc='upper left')

plt.savefig('vis/auto_arima_pred_vs_test.png', bbox_inches="tight")
plt.show()
```

**y_avg_train**

```python
# Fit a simple auto_arima model
mod_avg_auto = auto_arima(y_avg_train, start_p=0, start_q=0,
                     max_p=10, max_q=10, d=0,
                     stepwise=True, suppress_warnings=True,
                     error_action='ignore')

mod_avg_auto.order
```

```python
mod_avg_auto.summary()
```

```python
# Create predictions for the future, evaluate on test
preds_avg, conf_int_avg = mod_avg_auto.predict(n_periods=y_avg_test.shape[0], return_conf_int=True)

# Print the error:
print("Test RMSE: %.3f" % np.sqrt(mean_squared_error(y_avg_test, preds_avg)))
```

```python
plt.figure(figsize=(14,7))

plt.plot(training_days, y_avg_train, alpha=0.75, label='Training Data')
plt.plot(testing_days, preds_avg, alpha=0.75, label='Predictions')  # Forecasts
plt.plot(testing_days, y_avg_test,
            alpha=0.4, linestyle='--', label='Actual/Test Data')  # Test data
plt.fill_between(testing_days,
                 conf_int_avg[:, 0], conf_int_avg[:, 1],
                 alpha=0.1, color='b')

plt.title("GDB forecasts using auto_arima")
plt.xlabel("Date")
plt.ylabel("Blood Glucose Reading (mg/dL)")
plt.legend(loc='upper left')

plt.savefig('vis/auto_arima_avg_pred_vs_test.png', bbox_inches="tight")
plt.show()
```
